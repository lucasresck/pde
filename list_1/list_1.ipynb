```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
```

```python
def euler(dy, y_0, delta_t, T):
    y = [y_0]
    for i, t in enumerate(np.arange(y_0, T, delta_t)):
        next_y = dy(y[i])*delta_t + y[i]
        y.append(next_y)
    return y
```

```python
def plot(dy, y, y_0, delta_y, T):
    sns.lineplot(np.arange(y_0, T+delta_t, delta_t), euler(dy, y_0, delta_t, T), label='Euler')
    sns.lineplot(np.arange(y_0, T+delta_t, delta_t), y(y_0, delta_t, T), label='Original')
    plt.title('Solution to ordinary differential equation')
    plt.show()
```

## Question 2

```python
y_0, delta_t, T = 0, 0.1, 1
```

### a)

```python
def dy_a(y):
    return y + 5
```

```python
def y_a(y_0, delta_t, T):
    return 5*np.e**np.arange(y_0, T+delta_t, delta_t) - 5
```

```python
plot(dy_a, y_a, y_0, delta_t, T)
```

### b)

```python
def dy_b(y):
    return -2*y + 5
```

```python
def y_b(y_0, delta_t, T):
    return 5/2 - 5/2*np.e**(-2*np.arange(y_0, T+delta_t, delta_t))
```

```python
plot(dy_b, y_b, y_0, delta_t, T)
```
